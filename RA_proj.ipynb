# Confidence Interval

## Size    

```r
n <- nrow(train_set)         # sample_size

beta_hat  <- 36.602
std_error <- 1.921

conf_level  <- 0.95                                 # = 95 %
percentile  <- conf_level + (1 - conf_level)/2      # = 97.5 %
t_value     <- qt(percentile, df=n-17)               # = 2.01

cutoff_lhs <- beta_hat - t_value*std_error
cutoff_rhs <- beta_hat + t_value*std_error

confidence_interval <- c(cutoff_lhs, cutoff_rhs)
print(confidence_interval)
```
